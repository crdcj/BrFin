```python
cols = petro._df_main.columns
df = petro._df_main[cols[3:-2]]
df
```

```python
last_afs = df.query('report_period == "annual"')['DT_FIM_EXERC'].max()
last_qfs = df.query('report_period == "quarterly"')['DT_FIM_EXERC'].max()
last_qfs > last_afs
```

```python
df1 = df.query('DT_FIM_EXERC == @last_qfs').copy()
df1.reset_index(drop=True, inplace=True)
df1
```

```python
df1.query('DT_INI_EXERC == DT_INI_EXERC.min()',
    inplace=True)
df1
```

```python
df2 = df.query('DT_REFER == @last_qfs').copy()
df2
```

```python
df2.query('DT_INI_EXERC == DT_INI_EXERC.min()', inplace=True)
df2['VL_CONTA'] = -df2['VL_CONTA']
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df3 = df.query('DT_FIM_EXERC == @last_afs').copy()
df3.reset_index(drop=True, inplace=True)
df3
```

```python
df_ltm = pd.concat([df1,df2,df3])
df_ltm
```

```python
df_ltm = df_ltm[['CD_CONTA', 'VL_CONTA']]
df_ltm
```

```python
df_ltm = df_ltm.groupby(by='CD_CONTA').sum().reset_index()
df_ltm
```
